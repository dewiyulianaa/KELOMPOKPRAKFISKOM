VISUALISASI DATA MENGGUNAKAN PLOT

```python
import matplotlib.pyplot as plt

# Baca file txt
file_path = r"D:\KULIAH\semester 5\Prak.Fiskom\gjb1.txt"
with open(file_path, 'r') as file:
    lines = file.readlines()

# Proses data
categories = []
values = []

for line in lines:
    # Misalnya, asumsikan bahwa setiap baris berisi dua nilai yang dipisahkan oleh koma
    category, value = line.strip().split(',')
    categories.append(category)
    values.append(float(value))

# Buat plot
plt.figure(figsize=(16, 5))

# Plot data dalam bentuk garis
plt.subplot(1, 4, 1)
plt.plot(values, marker='o', linestyle='-', color='b')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot')

# Plot histogram
plt.subplot(1, 4, 2)
plt.hist(values, bins=20, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Plot bar
plt.subplot(1, 4, 3)
plt.bar(categories, values, color='green')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')

# Plot scatter
plt.subplot(1, 4, 4)
plt.scatter(range(len(values)), values, color='red', marker='x')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Scatter Plot')

# Menyimpan plot secara otomatis
output_file_path = 'output_plot.png'
plt.savefig(output_file_path)

# Menampilkan plot
plt.tight_layout()
plt.show()

print(f"Plot telah disimpan sebagai {output_file_path}")

```
